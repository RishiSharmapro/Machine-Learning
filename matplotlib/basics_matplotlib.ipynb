```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ploting a basic graph

x = list(range(8))
y = [4, 6, 1, 3, 2, 5, 8, 1]

plt.plot(x, y)
plt.show()
```

```python
# simple line graph for a store 

x = ["Mon", "Tues", "Wed", "Thur", "Fri"]
y = [12, 14, 10, 20, 25]

plt.plot(x, y)

plt.title('Bakery sales - current week')
plt.xlabel('Day of the week')
plt.ylabel('Sales per day')

plt.show()
```
