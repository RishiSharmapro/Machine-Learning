```python
import matplotlib.pyplot as plt
```

```python
# plt.plot(x, y, color='name/hex', linestyle='dash/dot/etc', linewidth=width, marker='marker_symbol', label='label name')
# marker can be <, >, ^, o, s(square), d(diamond), x, + etc

months = [1, 2, 3, 4, 5]
sales = [1000, 1200, 1500, 1900, 2200]

plt.plot(months, sales, color='blue', linestyle='--', linewidth=2, marker='o', label='2025 sales data')
plt.xlabel('Months') # set x label
plt.ylabel('Sales in month') # set y label
plt.title('Monthly Sales Data Report') # set title of the graph
plt.legend(loc='upper left', fontsize=9) # for this the plot should have a label feild or can pass a string but show only first char
plt.grid(color='gray', linestyle=':', linewidth=1)
plt.xlim(0.9, 5)
plt.ylim(900, 3000)
plt.xticks([1, 2, 3, 4, 5, 6], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])
plt.show()
```

```python
# plt.bar(x, height, color='name/hex',  width=value, label='label name')

product = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 800, 1200]

plt.bar(product, sales, color='orange', label='Product Report') # for vertical bar charts
# plt.barh(product, sales, color='orange', label='Product Report') # for horizontal bar charts
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.legend()
plt.show()
```

```python
# plt.pie(values, label=label_list, colors=colors_list, autopct='%1.1f%%'/'%1d%%')

regions = ['North', 'South', 'East', 'West']
revenue = [100, 40, 50, 10]

plt.pie(revenue, labels=regions, colors=['gold', 'skyblue', 'lightgreen', 'coral'], autopct='%1d%%')
plt.title('Revenue Divison among Regions')
plt.show()
```

```python
# plt.hist(data,, bins=number_of_bins, color='name/hex', edgecolor='name/hex') bins = range

scores = [40, 56, 69, 57, 89, 90, 95, 85, 84, 91, 77, 74, 62, 55, 51, 84, 79, 47, 81]

plt.hist(scores, bins=5, color='green', edgecolor='black')
plt.xlabel('Scores of student')
plt.ylabel('Number of student')
plt.title('Result Analysis os students')
plt.show()
```

```python
# plt.scatter(x, y, color'color_name', marker='marker_style', label='label name')

hours_studied = [1, 2, 3, 4, 5, 6, 7]
markes_scored = [55, 67, 78, 70, 85, 90, 87]

plt.scatter(hours_studied, markes_scored, color='purple', marker='x', label='Student Hour-to-Score Analysis')
plt.xlabel('Hour Studied')
plt.ylabel('Markes Scored')
plt.title('Relationship between Hours Studied and Markes Scored')
plt.legend()
plt.grid()
plt.show()
```

```python
# using multiple groups in same scatter plot 

hours_studied = [1, 2, 3, 4, 5, 6, 7]

markes_scored_grpA = [55, 67, 78, 70, 85, 90, 87]
markes_scored_grpB = [69, 75, 85, 90, 85, 80, 82]

plt.scatter(hours_studied, markes_scored_grpA, color='purple', marker='x', label='Student Hour-to-Score Analysis')
plt.scatter(hours_studied, markes_scored_grpB, color='blue', marker='o', label='Student Hour-to-Score Analysis')
plt.xlabel('Hour Studied')
plt.ylabel('Markes Scored')
plt.title('Relationship between Hours Studied and Markes Scored')
plt.legend(loc='lower right', fontsize=8)
plt.grid()
plt.show()
```

```python
# way 1: using subplot
# plt.subplot(nrows, ncols, index)

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.subplot(1, 2, 1) # row 1, col 2, subplot 1
plt.plot(x, y)
plt.title('Line Chart')

plt.subplot(1, 2, 2) # row 1, col 2, subplot 2
plt.bar(x, y)
plt.title('Bar Chart')

plt.tight_layout() # adjusts the spacing between subplots to make sure that nothing overlapes and make automatic adjustments
plt.show()
```

```python
# way 2: subplots
# fig, ax = plt.subplots(nrows, ncols, figsize={weight, height})

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

ax[0].plot(x, y, color='green')
ax[0].set_title('Line Plot')

ax[1].bar(x, y, color='orange')
ax[1].set_title('Bar Chart')

fig.suptitle('Comparision of Line and Bar Charts') # main title of the chart
# use of manual adjustment unlike tight_layout()
# fig.subplots_adjust(left=None, right=None, top=None, bottom=None, wspace=None, hspace=None)
# fig.subplots_adjust(left=0.2, right=0.8, top=0.9, bottom=0.1, wspace=0.4, hspace=0.6)

plt.tight_layout()
plt.show()
```

### The plt.savefig() function saves the current figure to a file (like .png, .jpg, .pdf, etc.).


| Parameter             | What it Does                                                       |
| --------------------- | ------------------------------------------------------------------ |
| `"filename"`          | Name and format (e.g. `'plot.pdf'`, `'graph.png'`)                 |
| `dpi=300`             | Sets image resolution (**dots per inch**) — higher = sharper       |
| `bbox_inches='tight'` | Crops extra whitespace around the figure                           |
| `transparent=True`    | Saves the background as **transparent** (great for logos/overlays) |
| `facecolor='color'`   | Sets background color of the saved figure                          |
| `edgecolor='color'`   | Sets edge color of the saved figure                                |
| `format='png'`        | Force save in a specific format regardless of filename             |


```python
# plt.savefig('filename.ext', dpi=value, bbox_inches='tight')

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o', label='Demo Sales Data')
plt.xlabel('Week')
plt.ylabel('Sales in $')
plt.title('Sales Analysis')
plt.grid()


plt.savefig('demo_sales.png', dpi=400, bbox_inches='tight', facecolor='lightblue')
plt.show() # remember it clears the plot after rendering so save before show()
```
